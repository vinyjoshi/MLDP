```python
import pandas as pd
from sklearn.preprocessing import LabelEncoder
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
```

```python
df = pd.read_csv('titanic.csv')
df.head(10)
```

```python
data = df.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis= 'columns')
data
```

```python
data = data.fillna(df.Age.median())
data.head(10)
```

```python
features = data.drop('Survived', axis="columns")
survived = data[['Survived']]
```

```python
le_sex = LabelEncoder()
features['sex_'] = le_sex.fit_transform(features[['Sex']])
features = features.drop('Sex', axis='columns')
features
```

```python
xTrain, xTest, yTrain, yTest = train_test_split(features, survived, test_size=0.2, random_state=10)
```

```python
model = DecisionTreeClassifier()
model.fit(xTrain, yTrain)
```

```python
model.score(xTest, yTest)
```

```python
model.predict(xTest)
```
